# Mô tả các thuộc tính
1. **Age**: tuổi của bệnh nhân
2. **Sex**: giới tính (nam/nữ)
3. **ChestPainType**: loại đau ngực (TA: Typical Angina, ATA: Atypical Angina, NAP: Non-Anginal Pain, ASY: Asymptomatic)
4. **RestingBP**:huyết áp nghỉ (đơn vị: mmHg)
5. **Cholesterol**: cholesterol trong huyết thanh (đơn vị: mg/dl)
6. **FastingBS**: đường huyết nhanh (1 nếu > 120 mg/dl, ngược lại: 0)
7. **RestingECG**: kết quả điện tâm đồ nghỉ (Normal, ST, LVH)
8. **MaxHR**: nhịp tim tối đa đạt được
9. **ExerciseAngina**: đau ngực do vận động gây ra (Y: có, N: không)
10. **Oldpeak**: Oldpeak = ST ([Giá trị số đo trong độ giảm])
11. **ST_Slope**: độ dốc của đoạn ST đỉnh do vận động tạo ra ([Up: Tăng, Flat: Bằng phẳng, Down: Giảm])
12. ***HeartDisease***: bệnh tim ([1: có, 0: không])

---

## Loại đau ngực:
1. typical angina:
Loại đau ngực có triệu chứng rõ ràng, bệnh nhân có thể có cảm giác đau ngực như một cảm giác nặng, ép ngực, nhức nhối ở vùng ngực.
2. atypical angina:
Loại đau ngực này có triệu chứng không rõ ràng, bệnh nhân không nhất thiết bị đau ở vùng ngực mà còn có thể đau ở hai vai, cánh tay, cổ, hay lưng.
3. non-anginal:
Loại đau ngực không phải do vấn đề tim mạch gây nên.
4. asymptomatic:
Không bị, không có các triệu chứng đau ngực nào.

---

## Kết quả điện tâm đồ nghỉ
1. Normal:
Bình thường, điện tâm đồ của trái tim ổn định
2. ST-T Abnormality: Bất thường ở đoạn ST-T:
Có sự biến đổi ở đoạn ST-T trên điện tâm đồ
3. LV Hypertrophy:
Có sự phì đại (tăng kích thước) của thất trái của tim

```python
# Vu Bao An

df['Cholesterol'].value_counts()

# Cholesterol có giá trị bằng 0 được xuất hiện 172 lần
# Như vậy đã có 172 điểm dữ liệu có giá trị Cholesterol = 0
```

```python
# Vu Bao An

# Để khắc phục vấn đề này, có thể thay thế giá trị clolesterol = 0 bằng giá trị trung bình (cột mean)
# Để giá trị clolesterol được thay thế chính xác hơn, sử dụng KNNImputer
# KNNImputer: được sử dụng để thay thế các giá trị còn thiếu trong dữ liệu (NaN)
# bằng cách sử dụng thuật toán K-Nearest Neighbors (KNN)

# Chuyển các giá trị 0 trong Clolesterol thành NaN
df['Cholesterol'].replace(0, np.nan, inplace=True)

from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=5) # lấy giá trị 5 hàng xóm gần nhất
replaced = imputer.fit_transform(df)
df = pd.DataFrame(replaced, columns=df.columns)
```

```python
# Vu Bao An

# Kiểm tra trong 'Cholesterol' còn giá trị NaN không
df['Cholesterol'].isna().sum()
```

```python
# Vu Bao An

# Thực hiện tương tự cho RestingBP
df['RestingBP'].replace(0, np.nan, inplace=True)
imputer = KNNImputer(n_neighbors=5) # lấy giá trị 5 hàng xóm gần nhất
replaced = imputer.fit_transform(df)
df = pd.DataFrame(replaced, columns=df.columns)
```

```python
# Nguyen Gia My

X_train, X_test, y_train, y_test = train_test_split(
    data.drop('HeartDisease', axis=1), # bỏ target
    data['HeartDisease'],
    test_size=0.2,
    random_state=42,
    stratify=data['HeartDisease']
)
```

```python
# Nguyen Gia My

inputs = tf.placeholder(tf.float32, shape=(None, X_train.shape[1]), name='inputs')
label = tf.placeholder(tf.float32, shape=(None, 2), name='labels')

# First Layer
hid1_size = 128
w1 = tf.Variable(tf.random_normal([hid1_size, X_train.shape[1]], stddev=0.01), name='w1')
b1 = tf.Variable(tf.constant(0.1, shape=(hid1_size, 1)), name='b1')
y1 = tf.nn.dropout(tf.nn.relu(tf.add(tf.matmul(w1, tf.transpose(inputs)), b1)), keep_prob=0.5)

# Second layer
hid2_size = 256
w2 = tf.Variable(tf.random_normal([hid2_size, hid1_size], stddev=0.01), name='w2')
b2 = tf.Variable(tf.constant(0.1, shape=(hid2_size, 1)), name='b2')
y2 = tf.nn.dropout(tf.nn.relu(tf.add(tf.matmul(w2, y1), b2)), keep_prob=0.5)

# Output layer
wo = tf.Variable(tf.random_normal([2, hid2_size], stddev=0.01), name='wo')
bo = tf.Variable(tf.random_normal([2, 1]), name='bo')
yo = tf.transpose(tf.add(tf.matmul(wo, y2), bo))
```

```python
# Vu Bao An

# Loss function and optimizer
lr = tf.placeholder(tf.float32, shape=(), name='learning_rate')
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=yo, labels=label))
optimizer = tf.train.GradientDescentOptimizer(lr).minimize(loss)

# Prediction
pred = tf.nn.softmax(yo)
pred_label = tf.argmax(pred, 1)
correct_prediction = tf.equal(tf.argmax(pred, 1), tf.argmax(label, 1))
accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
```

```python
# Nguyen Gia My

# Train
# Create operation which will initialize all variables
init = tf.global_variables_initializer()

# Configure GPU not to use all memory
config = tf.ConfigProto()
config.gpu_options.allow_growth = True

# Start a new tensorflow session and initialize variables
sess = tf.InteractiveSession(config=config)
sess.run(init)
```
